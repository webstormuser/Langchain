# Making retriever

```python
#making vectordb as retriever
vectordb = Chroma(persist_directory=persist_directory, embedding_function=embeddings)
retriever = vectordb.as_retriever()
retriever
```

```python
# getting relevant_documents
# Example query
query = "What is the capital of France?"

# Retrieve the most relevant documents
relevant_documents = retriever.get_relevant_documents(query, top_k=3)

# Display the results
for i, doc in enumerate(relevant_documents):
    # Access the content using the "page_content" attribute instead of "text"
    print(f"Document {i+1}: {doc.page_content}")
```

```python
# getting relevant_documents
# Example query
query = "I love the APPLE"

# Retrieve the most relevant documents
relevant_documents = retriever.get_relevant_documents(query, top_k=3)

# Display the results
for i, doc in enumerate(relevant_documents):
    # Access the content using the "page_content" attribute instead of "text"
    print(f"Document {i}: {doc.page_content}")
```

```python
# Retrieve the most relevant documents
query = " I like Nokiya"
results = retriever.get_relevant_documents(query, top_k=3,search_kwargs={"k":3})

# Display the results
for i, result in enumerate(results):
    # Access the content using the "page_content" attribute
    print(f"Result {i+1}: {result.page_content}")

# Display the query
print(f"Query: {query}")
```
